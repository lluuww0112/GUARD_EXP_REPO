```python
from huggingface_hub import snapshot_download

# MVBench 데이터셋 ID
repo_id = "OpenGVLab/MVBench"

# 구글 드라이브 내 저장 경로
save_path = "/content/drive/MyDrive/MVBench_Dataset"

# 다운로드 시작
snapshot_download(
    repo_id=repo_id,
    repo_type="dataset",
    local_dir=save_path,
    resume_download=True
)
```
